#Dados ausentes
Métodos convenientes para lidar com Missing Data em pandas:

O método "dropna" elimina todas as linhas que tenham algum atributo com valor vazio

```
idh_aux = idh.dropna()
```

Note que ao verificar as novas informações da tabela com info() idh_aux tem 4491 linhas (antes do drop eram 5508 entradas).

```
idh_aux.info()
```

#Preenchendo vazios com valores
Também é possível substituir os valores VAZIOS por outros valores automaticamente. Para tal, pode-se usar o método "fillna"

```
idh.fillna(value=0)
```
